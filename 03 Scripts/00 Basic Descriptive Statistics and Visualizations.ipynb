### 03.1 Calculate the Number of Outliers

```python
# Calculate Interquartile Range (IQR)

q1_floor_area_sqm = np.quantile(flat_resale['floor_area_sqm'], 0.25)
q3_floor_area_sqm = np.quantile(flat_resale['floor_area_sqm'], 0.75)
iqr_floor_area_sqm = q3_floor_area_sqm - q1_floor_area_sqm
print("Q1 of 'floor_area_sqm': {}".format(q1_floor_area_sqm))
print("Q3 of 'floor_area_sqm': {}".format(q3_floor_area_sqm))
print("IQR of 'floor_area_sqm': {}".format(iqr_floor_area_sqm))
print()

q1_remaining_lease = np.quantile(flat_resale['remaining_lease'], 0.25)
q3_remaining_lease = np.quantile(flat_resale['remaining_lease'], 0.75)
iqr_remaining_lease = q3_remaining_lease - q1_remaining_lease
print("Q1 of 'remaining_lease': {}".format(q1_remaining_lease))
print("Q3 of 'remaining_lease': {}".format(q3_remaining_lease))
print("IQR of 'remaining_lease': {}".format(iqr_remaining_lease))
print()

q1_station_distance = np.quantile(flat_resale['station_distance'], 0.25)
q3_station_distance = np.quantile(flat_resale['station_distance'], 0.75)
iqr_station_distance = q3_station_distance - q1_station_distance
print("Q1 of 'station_distance': {}".format(q1_station_distance))
print("Q3 of 'station_distance': {}".format(q3_station_distance))
print("IQR of 'station_distance': {}".format(iqr_station_distance))
print()

q1_storey_range_midpoint = np.quantile(flat_resale['storey_range_midpoint'], 0.25)
q3_storey_range_midpoint = np.quantile(flat_resale['storey_range_midpoint'], 0.75)
iqr_storey_range_midpoint = q3_storey_range_midpoint - q1_storey_range_midpoint
print("Q1 of 'storey_range_midpoint': {}".format(q1_storey_range_midpoint))
print("Q3 of 'storey_range_midpoint': {}".format(q3_storey_range_midpoint))
print("IQR of 'storey_range_midpoint': {}".format(iqr_storey_range_midpoint))
print()

q1_cpi_all_items = np.quantile(flat_resale['cpi_all_items'], 0.25)
q3_cpi_all_items = np.quantile(flat_resale['cpi_all_items'], 0.75)
iqr_cpi_all_items = q3_cpi_all_items - q1_cpi_all_items
print("Q1 of 'cpi_all_items': {}".format(q1_cpi_all_items))
print("Q3 of 'cpi_all_items': {}".format(q3_cpi_all_items))
print("IQR of 'cpi_all_items': {}".format(iqr_cpi_all_items))
print()

q1_resale_price = np.quantile(flat_resale['resale_price'], 0.25)
q3_resale_price = np.quantile(flat_resale['resale_price'], 0.75)
iqr_resale_price = q3_resale_price - q1_resale_price
print("Q1 of 'resale_price': {}".format(q1_resale_price))
print("Q3 of 'resale_price': {}".format(q3_resale_price))
print("IQR of 'resale_price': {}".format(iqr_resale_price))
print()

q1_adjusted_resale_price = np.quantile(flat_resale['adjusted_resale_price'], 0.25)
q3_adjusted_resale_price = np.quantile(flat_resale['adjusted_resale_price'], 0.75)
iqr_adjusted_resale_price = q3_adjusted_resale_price - q1_adjusted_resale_price
print("Q1 of 'adjusted_resale_price': {}".format(q1_adjusted_resale_price))
print("Q3 of 'adjusted_resale_price': {}".format(q3_adjusted_resale_price))
print("IQR of 'adjusted_resale_price': {}".format(iqr_adjusted_resale_price))
```

```python
# Calculate values of upper and lower whisker

lower_whisker_floor_area_sqm = q1_floor_area_sqm - (1.5 * iqr_floor_area_sqm)
upper_whisker_floor_area_sqm = q3_floor_area_sqm + (1.5 * iqr_floor_area_sqm)

lower_whisker_remaining_lease = q1_remaining_lease - (1.5 * iqr_remaining_lease)
upper_whisker_remaining_lease = q3_remaining_lease + (1.5 * iqr_remaining_lease)

lower_whisker_station_distance = q1_station_distance - (1.5 * iqr_station_distance)
upper_whisker_station_distance = q3_station_distance + (1.5 * iqr_station_distance)

lower_whisker_storey_range_midpoint = q1_storey_range_midpoint - (1.5 * iqr_storey_range_midpoint)
upper_whisker_storey_range_midpoint = q3_storey_range_midpoint + (1.5 * iqr_storey_range_midpoint)

lower_whisker_cpi_all_items = q1_cpi_all_items - (1.5 * iqr_cpi_all_items)
upper_whisker_cpi_all_items = q3_cpi_all_items + (1.5 * iqr_cpi_all_items)

lower_whisker_resale_price = q1_resale_price - (1.5 * iqr_resale_price)
upper_whisker_resale_price = q3_resale_price + (1.5 * iqr_resale_price)

lower_whisker_adjusted_resale_price = q1_adjusted_resale_price - (1.5 * iqr_adjusted_resale_price)
upper_whisker_adjusted_resale_price = q3_adjusted_resale_price + (1.5 * iqr_adjusted_resale_price)
```

```python
# Number of outliers based on 1.5*IQR

outlier_floor_area_sqm = len(flat_resale.loc[(flat_resale['floor_area_sqm'] < lower_whisker_floor_area_sqm) | 
                                             (flat_resale['floor_area_sqm'] > upper_whisker_floor_area_sqm)])

outlier_remaining_lease = len(flat_resale.loc[(flat_resale['remaining_lease'] < lower_whisker_remaining_lease) | 
                                              (flat_resale['remaining_lease'] > upper_whisker_remaining_lease)])

outlier_station_distance = len(flat_resale.loc[(flat_resale['station_distance'] < lower_whisker_station_distance) | 
                                               (flat_resale['station_distance'] > upper_whisker_station_distance)])

outlier_storey_range_midpoint = len(flat_resale.loc[(flat_resale['storey_range_midpoint'] < lower_whisker_storey_range_midpoint) | 
                                                    (flat_resale['storey_range_midpoint'] > upper_whisker_storey_range_midpoint)])

outlier_cpi_all_items = len(flat_resale.loc[(flat_resale['cpi_all_items'] < lower_whisker_cpi_all_items) | 
                                            (flat_resale['cpi_all_items'] > upper_whisker_cpi_all_items)])

outlier_resale_price = len(flat_resale.loc[(flat_resale['resale_price'] < lower_whisker_resale_price) | 
                                           (flat_resale['resale_price'] > upper_whisker_resale_price)])

outlier_adjusted_resale_price = len(flat_resale.loc[(flat_resale['adjusted_resale_price'] < lower_whisker_adjusted_resale_price) | 
                                                    (flat_resale['adjusted_resale_price'] > upper_whisker_adjusted_resale_price)])

print("No of outlier for 'floor_area_sqm': {}".format(outlier_floor_area_sqm))
print("No of outlier for 'remaining_lease': {}".format(outlier_remaining_lease))
print("No of outlier for 'station_distance': {}".format(outlier_station_distance))
print("No of outlier for 'storey_range_midpoint': {}".format(outlier_storey_range_midpoint))
print("No of outlier for 'cpi_all_items': {}".format(outlier_cpi_all_items))
print("No of outlier for 'resale_price': {}".format(outlier_resale_price))
print("No of outlier for 'adjusted_resale_price': {}".format(outlier_adjusted_resale_price))
```

The number of outliers are recorded in Excel report.
