### 04.1 Adjusted Resale Price

The Consumer Price Index (CPI) is widely used as a measure of the consumer price inflation. Inflation itself would diminish the purchasing power of a currency, effectively erroding its value over time. <br>
Given that the primary dataset on resale flat prices covers a long timeframe (from Jan 1990 to Dec 2023), the CPI values could be used to standardize the prices against the fluctuating currency's value across different periods. This allows the price comparison from different time periods on a more equitable basis. 

```python
# Create the new column 'adjusted_resale_price' using the formula below
flat_resale_cpi['adjusted_resale_price'] = flat_resale_cpi['resale_price'] / flat_resale_cpi['cpi_all_items'] * 100
```

```python
# Check if there is any missing data in the new column
flat_resale_cpi['adjusted_resale_price'].isnull().sum()
```

### 04.3 Interchange Flag - Is the nearest station an interchange?

```python
# 1) Create a list of station interchange

# Extract the station names that appear more than once in 'train_station' dataframe into a list
interchange_list = train_station.loc[train_station['station_name'].duplicated()]['station_name'].to_list()
```

```python
len(interchange_list)
```

```python
interchange_list.sort()
```

```python
interchange_list
```

The station interchange list above contains duplicate values. 

```python
# Remove duplicates from 'interchange_list'
interchange_list = list(set(interchange_list))
```

```python
interchange_list.sort()
```

```python
len(interchange_list)
```

```python
interchange_list
```

```python
# 2) Create a list of dates when the stations above become an interchange, which is the second opening date of the stations

date_become_interchange_list = []

for station in interchange_list: 
    date = pd.to_datetime('1800-01-01')
    date_list = train_station.loc[train_station['station_name'] == station, 'opening'].to_list()
    date_list.sort()
    
    for i in range(2): 
        if date_list[i] > date: 
            date = date_list[i]
    date_become_interchange_list.append(date)
```

```python
len(date_become_interchange_list)
```

```python
# 3) Create a dataframe 'df_interchange' that store the station interchange name and date become interchange

df_interchange = pd.DataFrame({'station_name' : interchange_list, 
                               'date_become_interchange' : date_become_interchange_list})
```

```python
df_interchange
```

```python
# Reduce the date granularity of 'date_become_interchange' from day (yyyy-mm-dd) to month (yyyy-mm-01) 
# because the date granularity in flat_resale['month'] is up to month only

# Extract the year and month from existing 'date_become_interchange' and store into new column 'date_become_interchange_yymm'
df_interchange['date_become_interchange_yymm'] = df_interchange['date_become_interchange'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
# Convert 'date_become_interchange_yymm' into datetime
df_interchange['date_become_interchange_yymm'] = pd.to_datetime(df_interchange['date_become_interchange_yymm'])
```

```python
df_interchange
```

```python
# 4) Merge 'train_station' with 'df_interchange' to get their respective 'station_name_w_code' and 'opening' dates

df_interchange_merged = df_interchange.merge(train_station, on = 'station_name', how = 'left', indicator = True)
```

```python
df_interchange_merged
```

```python
len(df_interchange_merged)
```

```python
df_interchange_merged.drop(columns = ['station_code', 'line', 'closure', 'latitude', 'longitude', 'geometry', 
                                      'closure_w_dummy', 'opening_yymm', 'closure_w_dummy_yymm', '_merge'], inplace = True)
```

```python
df_interchange_merged
```

```python
# 5) Extract the flat resale records where their nearest stations is listed in df_interchange_merged['station_name_w_code']

flat_resale_interchange = flat_resale_cpi.loc[flat_resale_cpi['nearest_station'].isin(df_interchange_merged['station_name_w_code'].tolist())]
```

```python
len(flat_resale_interchange)
```

```python
flat_resale_interchange.head()
```

```python
# 6) Determine if the nearest station in 'flat_resale_interchange' has become an interchange in their transaction 'month'

interchange_flag_list = []

for i, row in flat_resale_interchange.iterrows(): 
    if (row['month'] >= df_interchange_merged.loc[df_interchange_merged['station_name_w_code'] == row['nearest_station']]['date_become_interchange_yymm']).item(): 
        interchange_flag_list.append('station interchange')
    else: 
        interchange_flag_list.append('normal station')
```

```python
len(interchange_flag_list)
```

```python
# Create a new column 'interchange_flag' for 'flat_resale_interchange' using 'interchange_flag_list'
flat_resale_interchange['interchange_flag'] = interchange_flag_list
```

```python
# Verify the changes
flat_resale_interchange['interchange_flag'].value_counts(dropna = False)
```

```python
flat_resale_interchange.head()
```

```python
# 7) Merge 'flat_resale_interchange' with the initial 'flat_resale_cpi'

flat_resale_merged = flat_resale_cpi.merge(flat_resale_interchange, on = ['month', 'row_id', 'nearest_station'], how = 'left')
```

```python
flat_resale_merged.head()
```

```python
flat_resale_merged['interchange_flag'].value_counts(dropna = False)
```

The records that have missing values in 'interchange_flag' column is normal because their nearest station is not inside the interchange list, and therefore not included in 'flat_resale_interchange'. They should all be normal station in this column. 

```python
# Fill the missing values in 'interchange_flag' column with value 'normal station'
flat_resale_merged['interchange_flag'].fillna('normal station', inplace = True)
```

### 04.4 storey_range_midpoint

The existing variable 'storey_range' is a quantitative variable. We could convert it to quantitative by assigning the midpoint value to each range. 

```python
flat_resale_merged['storey_range'].value_counts(dropna = False).sort_index()
```

```python
# 1) Create a list that contains the distinct values of storey range
storey_range_list = list(set(flat_resale_merged['storey_range'].to_list()))
storey_range_list.sort()
storey_range_list
```

```python
# 2) Create a list that contains the average value of each storey range

storey_range_midpoint_list = [2, 5, 8, 11, 14, 17, 20, 23, 26, 29, 32, 35, 38, 41, 44, 47, 50] 
```

```python
# 3) Create a dataframe using the two lists
df_storey_range_midpoint = pd.DataFrame({'storey_range' : storey_range_list, 
                                    'storey_range_midpoint' : storey_range_midpoint_list})
```

```python
df_storey_range_midpoint
```

```python
# 4) Merge 'df_storey_range_midpoint' with 'flat_resale_merged_2'

flat_resale_merged_2 = flat_resale_merged.merge(df_storey_range_midpoint, on = 'storey_range', how = 'left', indicator = True)
```

```python
# 5) Verify the changes

flat_resale_merged_2.head()
```

```python
flat_resale_merged_2['_merge'].value_counts(dropna = False)
```

```python
# Verify the min and max of the midpoint values for each storey range
flat_resale_merged_2.groupby(['storey_range']).agg({'storey_range_midpoint': ['min', 'max']})
```

```python
# 6) Remove the column '_merge'

flat_resale_merged_2.drop(columns = ['_merge'], inplace = True)
```
